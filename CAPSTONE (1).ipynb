```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

## MOBILES CATEGORY SCRAPING

```python
options=Options()
options.add_argument("--disable-gpu")
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=options)
driver.get("https://www.flipkart.com/search?q=mobiles%205g&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
driver.maximize_window()
wait=WebDriverWait(driver,10)
```

```python
ProductName=[]
Price=[]
Category=[]
Ratings=[]
NumberofReviews=[]
```

```python
base_url = "https://www.flipkart.com/search?q=mobiles%205g&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"

for page in range(1,7):
    driver.get(f"{base_url}&page={page}")
    print(f"Scraping page {page}...")

    wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "jIjQ8S")))
    mobiles = driver.find_elements(By.CLASS_NAME, "jIjQ8S")

    for mobile in mobiles:
        try:
            cate=driver.find_elements(By.CLASS_NAME,"aDN3s4")
            category = cate.find_element(By.CLASS_NAME, "zEfFJL").text.strip()
        except:
            category = "N/A"

        try:
            name = mobile.find_element(By.CLASS_NAME, "RG5Slk").text.strip()
        except:
            name = "N/A"

        try:
            price = mobile.find_element(By.CSS_SELECTOR, ".hZ3P6w.DeU9vF").text.strip()
        except:
            price = "N/A"

        try:
            rating = mobile.find_element(By.CLASS_NAME, "MKiFS6").text.strip()
        except:
            rating = "N/A"

        try:
            text = mobile.find_element(By.CLASS_NAME, "PvbNMB").text
            reviewcount = text.split("&")[1].replace("Reviews", "").replace(",", "").strip()
        except:
            reviewcount = "N/A"

        ProductName.append(name)
        Price.append(price)
        Category.append(category)
        Ratings.append(rating)
        NumberofReviews.append(reviewcount)

   
    
```

```python
#create DataFrame
df = pd.DataFrame({
    "ProductName": ProductName,
    "Price": Price,
    "Category": Category,
    "Rating": Ratings,
    "Reviews":NumberofReviews
})

print(f"\nTotal  scraped: {len(df)}")
```

```python
df['Category']="Mobile"
```

```python
df.head(10)
```

```python
driver.quit()
```

## ELECTRONICS CATEGORY SCRAPING
